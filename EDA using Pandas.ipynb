```python
ax = df["Year_Introduced"].value_counts().head(10).plot(kind = "bar", title = "Top 10 Coasters Introduced")

ax.set_label("Year Introduced")
ax.set_ylabel("Count")
plt.show()
```

```python
ax = df["Speed_mph"].plot(kind = "hist", bins = 30, title = "Coaster Speed mph")
ax.set_label("Speed (mph)")
```

```python
ax = df["Speed_mph"].plot(kind = "kde", title =  "Coaster Speed mph")
ax.set_xlabel("Speed (mph)")
plt.show()
```

```python
df.plot(kind = 'scatter',
        x = 'Speed_mph',
        y = 'Height_ft',
        title = 'Coaster Speed vs. Height')
plt.show()
```

```python
ax = sns.scatterplot(x = 'Speed_mph',
                y = 'Height_ft',
                hue = 'Year_Introduced',
                data = df)
ax.set_title('Coaster Speed vs. Height')
plt.show()
```

```python
sns.pairplot(df,
             vars=['Year_Introduced','Speed_mph',
                   'Height_ft','Inversions','Gforce'],
            hue='Type_Main')
plt.show()
```

```python
df_corr = df[['Year_Introduced', 'Speed_mph',
    'Height_ft', 'Inversions', 'Gforce']].dropna().corr()
df_corr
```

```python
sns.heatmap(df_corr, annot = True)
```

```python
ax = df.query('Location != "Other"') \
        .groupby('Location')['Speed_mph'] \
        .agg(['mean', 'count']) \
        .query('count >= 10') \
        .sort_values('mean')["mean"] \
        .plot(kind = "barh", figsize = (10, 5), title = "Avg. Coast Speed by Location")
ax.set_xlabel("Avg. Speed of the Coast")
plt.show()
```
